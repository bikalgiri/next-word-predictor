```python
with open("tech_data.txt") as f:
    contents = f.read()
```

```python
contents
```

```python

#text preprocessing
import re #importing regex
import string,time
from nltk.corpus import stopwords



contents=contents.lower() #converting the words into lowercase

def remove_html_tags(text): #remove all the tags,romans,\n.""etc
    cleaned = re.sub(
        r"<[^>]*>|\r?\n|\ufeff|\b[ivxlcdm]+\b|['\"]+",
        "",
        text,
        flags=re.IGNORECASE
    )
    cleaned = re.sub(r"\s{2,}", " ", cleaned).strip()
    return cleaned


def remove_url(text):   #function using regex to remove urls
  pattern=re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)

def remove_punctuations(text): #function to remove punctuations
  translator=str.maketrans('','',string.punctuation)
  return text.translate(translator)

def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return ' '.join(x)

```

```python
import nltk
nltk.download('stopwords')
content=remove_html_tags(contents)
content=remove_url(content)
content=remove_punctuations(content)
content=remove_stopwords(content)
```

```python
content
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer #tokenization can be done by spacy.stemming and lemmitization should be done.
```

```python
tokenizer=Tokenizer()
```

```python
#to convert the whole text into tokens of word
tokenizer.fit_on_texts([contents])
```

```python

#to see which token.no is assigned to each word
len(tokenizer.word_index)
```

```python
input_sequences=[]
for sentence in contents.split('\n'):
  tokenized_word=tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_word)):
    input_sequences.append(tokenized_word[:i+1])

```

```python
input_sequences
```

```python
#to find out the max length of sequences inside text for zero padding.
max_length=max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_length,padding='pre')
```

```python
x=padded_input_sequences[:,:-1]
```

```python
y=padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
#performing one-hot encoding to y
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Input
```

```python
model = Sequential([
    Input(shape=(max_length - 1,)),
    Embedding(84, 100),
    LSTM(150,return_sequences=True),
    LSTM(150),
    Dense(84, activation='softmax')
])

```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=70,batch_size=128,validation_split=0.2)
```

```python
import numpy as np
text="data"
for i in range(10):
  tokenized_text=tokenizer.texts_to_sequences([text])
  padded_text=pad_sequences(tokenized_text,maxlen=max_length-1,padding='pre')
  pos=np.argmax(model.predict(padded_text))

  for word,index in tokenizer.word_index.items():
    if index==pos:
      text=text+" "+word
      break
print(text)
```
